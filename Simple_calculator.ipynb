```python
import ipywidgets as widgets
from IPython.display import display

# Display box (like calculator screen)
display_box = widgets.Text(value='', disabled=True, layout=widgets.Layout(width='320px', height='50px'))

# Buttons layout (rows like a mobile calculator)
buttons = [
    ["7", "8", "9", "/"],
    ["4", "5", "6", "*"],
    ["1", "2", "3", "-"],
    ["0", ".", "=", "+"],
    ["C", "(", ")", "%"]
]

# Handle button clicks
def on_button_click(b):
    text = b.description
    if text == "=":
        try:
            result = str(eval(display_box.value))
            display_box.value = result
        except:
            display_box.value = "Error"
    elif text == "C":
        display_box.value = ""
    else:
        display_box.value += text

# Create button widgets dynamically
button_widgets = []
for row in buttons:
    row_buttons = []
    for text in row:
        btn = widgets.Button(description=text, layout=widgets.Layout(width="70px", height="60px"))
        btn.on_click(on_button_click)
        row_buttons.append(btn)
    button_widgets.append(widgets.HBox(row_buttons))

# Show calculator
display(display_box)
for row in button_widgets:
    display(row)
```
